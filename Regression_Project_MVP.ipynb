# GHG Emissions Regression Project MVP
Samuel Robbins

The goal of this project is to perform a linear regression on Greenhouse Gas (GHG) emission and company/facility information data to predict which industries, fuel types, unit types, etc. are the most important predictors of total GHG emissions. To begin

The target for my model is `GHG_total` and I have created dummy variables for: 
- `State_x`
- `Industry Type (sectors)_x`
- `Unit Type`
- `General Fuel Type`
- `PARENT COMPANY NAME`

The dummy variables have amounted to 9960 columns of data with individual feature sets which is causing the model to take a significant amount of time to run; as of this writing it is still running. 

I will pare down the feature classes and perhaps run individual regressions for each dummy variable to see if there is any feature set that is more strongly linked to total GHG emissions.

**NEXT STEPS**

The next steps I will take in this analysis will be to link company financial data to determine if there are certain revenue/expenses/equity/etc. amounts that correlate to higher GHG emissions. With this data I can predict if there is a certain threshold that would indicate a facility is/will become a significant contributor of GHG's.
